# Scale Invariant Feature Transform (SIFT)

[Additional reading](https://docs.opencv.org/4.x/da/df5/tutorial_py_sift_intro.html)

```python
image = cv.imread('data/images/cameraman.png', cv.IMREAD_GRAYSCALE)

sift = cv.SIFT_create(nfeatures = 150)          # 150 = number of features/keypoints
kp = sift.detect(image, None)                   # detecting keypoints
des = sift.compute(image, kp)                   # computing the descriptor
#kp, des = siftdetectAndCompute(image, None)    # you can also do both at once

plt.figure(figsize = (10,10))
img = cv.drawKeypoints(image, kp, -1, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)  # drawing keypoints
#          cv.drawKeypoints(image, keypoints, outputImage, color, flags)
#                   color = -1 --> automatically choose the color for drawing the keypoints
#                   flags = DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS --> keypoints will be drawn as circles with their size & orientation
plt.imshow(img)
```

# SIFT Feature Matching

```python
# Load images
image1 = cv.imread('data/images/cartoon1.pgm', cv.IMREAD_GRAYSCALE)
image2 = cv.imread('data/images/cartoon2.pgm', cv.IMREAD_GRAYSCALE)

# Initialize ORB detector
feat = cv.SIFT_create(nfeatures = 150)

# Find the keypoints and descriptors with ORB
keypoints1, descriptors1 = feat.detectAndCompute(image1, None)
keypoints2, descriptors2 = feat.detectAndCompute(image2, None)

# Create BFMatcher (Brute Force Matcher)
bf = cv.BFMatcher()

# Match descriptors
matches = bf.match(descriptors1, descriptors2)

# Sort them in ascending order of distance 
matches = sorted(matches, key=lambda x: x.distance)                                     # key --> what do we want to match based on?

# Draw matches
img_matches = cv.drawMatches(image1, keypoints1, image2, keypoints2, matches[:50], 
                              cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)                # drawing only the first 50 matches
#             cv.drawMatches(img1, keypoints1, img2, keypoints2, matches1to2, outImg, matchColor, singlePointColor, matchesMask, flags)
#                     matchColor --> color to draw the line connecting matched keypoints
#                     singlePointColor --> color to draw keypoints that don't have a match
#                     matchesMask --> specifies which matches should be drawn
                              
# Plot the matches using Matplotlib
plt.figure(figsize = (12,8))
plt.imshow(img_matches)
plt.title('SIFT Feature Matching')
plt.show()
```
